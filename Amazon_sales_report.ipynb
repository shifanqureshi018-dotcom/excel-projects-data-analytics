```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

import Amazon dataset

```python
df=pd.read_csv('/content/Amazon Sale Report.csv', engine='python')
df.head()
```

```python
#clean useless data
df.drop(columns=['promotion-ids', 'Unnamed: 22'], inplace=True)
```

```python
# missing value fix
df = df.dropna(subset=['Amount'])
```

```python
df['currency'] = df['currency'].fillna('INR')


```

```python
df['ship-city'] = df['ship-city'].fillna('Unknown')
df['ship-state'] = df['ship-state'].fillna('Unknown')
df['ship-postal-code'] = df['ship-postal-code'].fillna('000000')
df['ship-country'] = df['ship-country'].fillna('Unknown')
df['Courier Status'] = df['Courier Status'].fillna('Not Available')
df['fulfilled-by'] = df['fulfilled-by'].fillna('Unknown')
df['currency'] = df['currency'].fillna('INR')
```

```python
df.isnull().sum()
```

CONVERT DATE COLUMN INTO INDIVIDUAL MONTH ,MONTH_NAME YEAR AND DAY


```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df['Day'] = df['Date'].dt.day
df['MonthName'] = df['Date'].dt.strftime('%b')
```

```python
# Group by Year + MonthName
monthly_sales = df.groupby(['Year', 'MonthName'])['Amount'].sum()

# Plot graph
monthly_sales.plot(kind='line', marker='X')
plt.title("Monthly Sales Trend")
plt.xlabel("Year, Month")
plt.ylabel("Sales Amount")
plt.grid(True)
plt.show()
```

```python
df.groupby('Category')['Amount'].sum().sort_values(ascending=False)

```

```python
# Category-wise Sales
cat_sales = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=cat_sales.values, y=cat_sales.index)
plt.title("Category-wise Total Sales")
plt.xlabel("Total Sales Amount")
plt.ylabel("Category")
plt.grid(True)
plt.show()
```

top saling SKU


```python
#group by Sku
df.groupby('SKU')['Qty'].sum().sort_values(ascending=False)

```

```python
df['SKU'].head(20)
```

```python
# Top 15 SKUs by Quantity Sold
top_sku = df.groupby('SKU')['Qty'].sum().sort_values(ascending=False).head(15)

plt.figure(figsize=(10,6))
sns.barplot(x=top_sku.values, y=top_sku.index)
plt.title("Top 15 Best Selling SKUs")
plt.xlabel("Total Quantity Sold")
plt.ylabel("SKU")
plt.grid(True)
plt.show()
```

```python
#state wise sales
df.groupby('ship-state')['Amount'].sum().sort_values(ascending=False)

```

```python
# State-wise Sales
state_sales = df.groupby('ship-state')['Amount'].sum().sort_values(ascending=False).head(15)

plt.figure(figsize=(10,6))
sns.barplot(x=state_sales.values, y=state_sales.index)
plt.title("Top 15 States by Total Sales")
plt.xlabel("Sales Amount")
plt.ylabel("State")
plt.grid(True)
plt.show()
```

```python
#order status distribution
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Status')
plt.title("Order Status Distribution")
plt.xlabel("Order Status")
plt.ylabel("Count of Orders")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
#courier status analysis
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Courier Status')
plt.title("Courier Status Distribution")
plt.xlabel("Courier Status")
plt.ylabel("Count of Orders")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Daily sales trend (Date-wise)
daily_sales = df.groupby(df['Date'].dt.date)['Amount'].sum()

plt.figure(figsize=(12,6))
daily_sales.plot(marker='o')
plt.title("Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Sales Amount")
plt.grid(True)
plt.show()
```

```python
# Group by date and sum the 'Amount'
daily_sales = df.groupby(df['Date'].dt.date)['Amount'].sum()
print(daily_sales)
```
